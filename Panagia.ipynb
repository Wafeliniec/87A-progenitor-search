Panagia et al. used HST photometry to identify stellar populations around SN1987A, but they did not have Gaia DR3 proper motions. Their key result is that SN1987A lies in/near a loose ∼12 Myr population within the local 30 pc field. Gaia analysis: among the Gaia-visible stars in and around this region, do any share the proper motion of SN1987A or the local LMC field?

Important limitation: Gaia only sees the brighter part of the population and misses many of the low-mass PMS stars Panagia identified with HST. Therefore, Gaia is not replacing Panagia’s population study; it is adding proper-motion information for the subset of stars Gaia can measure.
